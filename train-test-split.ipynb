```python
DATAPATH = "data/train"
COVID_PATH = DATAPATH + "/COVID"
NORMAL_PATH = DATAPATH + "/Normal"
```

```python
covid_img_paths = []
for dirname, _, filenames in os.walk(COVID_PATH):
    for filename in filenames:
        if (filename[-3:] == 'png'):
            covid_img_paths.append(os.path.join(dirname, filename))

normal_img_paths = []
for dirname, _, filenames in os.walk(NORMAL_PATH):
    for filename in filenames:
        if (filename[-3:] == 'png'):
            normal_img_paths.append(os.path.join(dirname, filename))
```

```python
covid_img_paths
```

```python
p_test = 0.2
N_covid = len(covid_img_paths)
N_normal = len(normal_img_paths)
idx_covid = np.random.default_rng(seed=123).choice(N_covid, int(N_covid * p_test), replace=False)
idx_normal = np.random.default_rng(seed=123).choice(N_normal, int(N_normal * p_test), replace=False)
```

```python
for img_path in [covid_img_paths[i] for i in idx_covid]:
  filename = img_path.split(os.path.sep)[-1]
  os.rename(img_path, "data/test/COVID/"+filename)
```

```python
for img_path in [normal_img_paths[i] for i in idx_normal]:
  filename = img_path.split(os.path.sep)[-1]
  os.rename(img_path, "data/test/Normal/"+filename)
```
